```python
import numpy as np
import matplotlib.pyplot as plt
file ="/content/drive/MyDrive/Machine-Learning-2023/olympic100m_men.txt"
datos = np.loadtxt(file, delimiter=',').T
x=datos[0]
y=datos[1]
plt.scatter(x,y)
slope = ((np.mean(x) * np.mean(y)) - np.mean(x * y)) / ((np.mean(x) ** 2) - np.mean(x ** 2))
intercept = np.mean(y) - slope * np.mean(x)

# Crear puntos para la línea de regresión
x_regression = np.linspace(min(x), max(x), 100)
y_regression = slope * x_regression + intercept

print(slope)
print(intercept)
# Graficar los puntos de datos y la línea de regresión
plt.scatter(x, y, label='Datos')
plt.plot(x_regression, y_regression, color='red', label='Regresión lineal')
plt.xlabel('Años de participación')
plt.ylabel('Tiempos')
plt.title('Regresión Lineal por Mínimos Cuadrados')
plt.legend()
plt.grid(True)
plt.show()
```

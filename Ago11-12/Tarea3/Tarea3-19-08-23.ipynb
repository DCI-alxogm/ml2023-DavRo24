```python
import numpy as np
import matplotlib.pyplot as plt
from sklearn.linear_model import LinearRegression

def w_mat1(x, y):
    X = np.vstack((np.ones_like(x), x, x**2)).T
    t = y
    XX = np.matmul(X.T, X)
    invXX = np.linalg.inv(XX)
    Xt = np.matmul(X.T, t)
    w = np.matmul(invXX, Xt)
    return w

def w_mat2(x, y):
    X = np.vstack((np.ones_like(x), x, x**2)).T
    t = y
    XX = np.matmul(X.T, X)
    invXX = np.linalg.inv(XX)
    Xt = np.matmul(X.T, t)
    w = np.matmul(invXX, Xt)
    return w

def w_mat3(x, y):
    X = np.vstack((np.ones_like(x), x, x**2)).T
    t = y
    Xt = np.matmul(X.T, t)
    XX_inv = np.linalg.inv(np.matmul(X.T, X))
    w = np.matmul(XX_inv, Xt)
    return w

# Cargando los datos
file = "/content/drive/MyDrive/Machine-Learning-2023/olympic100m_men.txt"
datos = np.loadtxt(file, delimiter=',').T
x = datos[0]
y = datos[1]

# Calculando la pendiente y la intersección para la regresión lineal
slope = ((np.mean(x) * np.mean(y)) - np.mean(x * y)) / ((np.mean(x) ** 2) - np.mean(x ** 2))
intercept = np.mean(y) - slope * np.mean(x)

# Crear puntos para la línea de regresión
x_regression = np.linspace(min(x), max(x), 100)
y_regression = slope * x_regression + intercept

# Graficar los puntos de datos y la línea de regresión
plt.scatter(x, y, label='Datos')
plt.plot(x_regression, y_regression, color='red', label='Regresión lineal')

# Ajustar y visualizar diferentes modelos usando mínimos cuadrados
w_2 = w_mat2(x, y)
y_regression_2 = w_2[2] * x_regression**2 + w_2[1] * x_regression + w_2[0]
plt.plot(x_regression, y_regression_2, color='green', label='Regresión cuadrática')

#Ajustando con biblioteca numpy
w_3 = w_mat3(x, y)
y_regression_3 = w_3[2] * x_regression**2 + w_3[1] * x_regression + w_3[0]
plt.plot(x_regression, y_regression_3, color='orange', label='Regresión cuadrática v2')

# Ajustar y visualizar el modelo usando scikit-learn
lr = LinearRegression()
x_1 = x.reshape(-1, 1)
lr.fit(x_1, y)
y_lr = lr.predict(x_1)
plt.plot(x, y_lr, color='blue', label='Regresión lineal sklearn')

plt.xlabel('Años de participación')
plt.ylabel('Tiempos')
plt.title('Regresión Lineal por Mínimos Cuadrados')
plt.legend()
plt.grid(True)
plt.show()

```
